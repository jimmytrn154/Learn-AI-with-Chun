```python
# %%
# IQR outlier share per numeric column
num_cols = data.select_dtypes(include=[np.number])
def outlier_share(s):
    q1,q3 = s.quantile([.25,.75])
    iqr = q3-q1
    return ((s < (q1-1.5*iqr)) | (s > (q3+1.5*iqr))).mean()
num_cols.apply(outlier_share).sort_values(ascending=False).head(20)
```

## Task 2 — Cleaning

**Strategy**
- Trimmed/squished text, normalized case for key categoricals; parsed numeric strings to numbers.
- Median imputed numeric NAs; added `"missing"` category to categorical fields.
- Dropped full row duplicates; removed NB classifier scores and plan to drop `Naive_Bayes_Classifier_` columns due to redundancy, which is also mentioned by the author of the dataset.
- Preserved the original file (`df_raw`) and worked on a copy (`df`) for reproducibility.

```python
obj_cols = df.select_dtypes(include="object").columns
for c in obj_cols:
    df[c] = df[c].astype(str).str.strip().replace({"": np.nan})

df["Attrition_Flag"] = df["Attrition_Flag"].astype("category")
df["Gender"] = df["Gender"].str.upper()
df["Education_Level"] = df["Education_Level"].str.title()

# Fix numeric types
fix_num = ["Customer_Age","Months_on_book","Dependent_count","Total_Relationship_Count",
           "Months_Inactive_12_mon","Contacts_Count_12_mon","Credit_Limit",
           "Total_Revolving_Bal","Avg_Open_To_Buy","Total_Trans_Amt","Total_Trans_Ct",
           "Total_Ct_Chng_Q4_Q1","Total_Amt_Chng_Q4_Q1","Avg_Utilization_Ratio"]
for c in fix_num:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Duplicates
df = df.drop_duplicates()

# Missing handling
num_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(include=["object","category"]).columns

df[num_cols] = df[num_cols].apply(lambda s: s.fillna(s.median()))
for c in cat_cols:
    df[c] = df[c].astype("category").cat.add_categories(["missing"]).fillna("missing")

# Drop leakage columns
leak_cols = [c for c in df.columns if c.startswith("Naive_Bayes_Classifier_")]
df = df.drop(columns=leak_cols, errors="ignore")
```

## Task 3 — Transformations

**Feature engineering logic**
- `inactivity_ratio` reflects disengagement risk; `trans_intensity` reflects engagement.
- `age_group` indicate better generalization for the dataset since the age varies from many sectors. Thus, group in to groups such as <=30, 31-40 will helps.
- Log transforms reduce skew for spend/limits; age groups aid reporting and monotonic behavior.


```python
# Age groups, ratios
df["age_group"] = pd.cut(df["Customer_Age"], bins=[0,30,40,50,60,200],
                         labels=["<=30","31-40","41-50","51-60","60+"], right=True)

df["inactivity_ratio"] = np.where(df["Months_on_book"]>0,
                                  df["Months_Inactive_12_mon"]/df["Months_on_book"], np.nan)
df["trans_intensity"]  = np.where(df["Months_on_book"]>0,
                                  df["Total_Trans_Ct"]/df["Months_on_book"], np.nan)

# Log transforms
for c in ["Total_Trans_Amt","Total_Trans_Ct","Credit_Limit"]:
    df[c+"_log"] = np.log1p(df[c])

# One-hot encoding (get_dummies)
ohe = pd.get_dummies(df[["Gender","Education_Level","Income_Category","Card_Category"]],
                     drop_first=False, dtype=int)
df_fe = pd.concat([df.drop(columns=["Gender","Education_Level","Income_Category","Card_Category"]),
                   ohe], axis=1)
```

## Task 4 — Aggregation & Final dataset

```python
# %%
by_age = (df_fe.groupby(["age_group","Attrition_Flag"])
          .agg(n=("CLIENTNUM","count"),
               avg_trans=("Total_Trans_Ct","mean"),
               avg_spend=("Total_Trans_Amt","mean"),
               util_med=("Avg_Utilization_Ratio","median"))
          .reset_index())
by_age = by_age.rename(columns={'n': 'client_count'})

by_age.head()
```

```python
# Churn rate by card category
ct = (df.groupby(["Card_Category","Attrition_Flag"])
        .size().rename("n").reset_index())
churn_by_card = (ct.assign(total=ct.groupby("Card_Category")["n"].transform("sum"),
                           churn_rate=lambda d: d["n"]/d["total"])
                   .pivot(index="Card_Category", columns="Attrition_Flag", values="churn_rate")
                   .reset_index())
churn_by_card.head()
```
