```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('data.xlsx', engine='openpyxl')
```

```python
df.head()
```

```python
count = df['subtype'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.countplot(x='subtype', data=df)
plt.title('Count of Each Subtype')
plt.xlabel('Subtype')
plt.ylabel('Count')
# plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```
