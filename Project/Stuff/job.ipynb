```python
import json
import os
from bs4 import BeautifulSoup

# ==============================================================================
# CONFIGURATION
# ==============================================================================
INPUT_FILE = 'nhatot_page.html'
OUTPUT_JSON = 'vinhomes_data.json'

def clean_text(text):
    if not text: return ""
    return text.strip()

def extract_html_to_json():
    if not os.path.exists(INPUT_FILE):
        print(f"Error: {INPUT_FILE} not found.")
        return

    print(f"Reading {INPUT_FILE}...")
    with open(INPUT_FILE, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    script_tag = soup.find('script', id='__NEXT_DATA__')
    if not script_tag:
        print("Error: Could not find __NEXT_DATA__ tag.")
        return

    try:
        raw_json = json.loads(script_tag.string)
        
        # --- ROBUST SEARCH STRATEGY ---
        # We look for the 'ads' list in likely locations
        ads_list = []
        
        # Path 1: Standard Page Props (Most common)
        try:
            ads_list = raw_json['props']['pageProps']['initialProps']['adlisting']['data']['ads']
        except KeyError:
            pass
            
        # Path 2: Server Side Props (Alternative)
        if not ads_list:
            try:
                ads_list = raw_json['props']['pageProps']['adlisting']['data']['ads']
            except KeyError:
                pass
                
        # Path 3: Search recursively if specific paths fail
        if not ads_list:
            print("Standard paths failed. Searching JSON tree for 'ads' key...")
            # Helper to find key in nested dict
            def find_key(obj, key):
                if key in obj: return obj[key]
                for k, v in obj.items():
                    if isinstance(v, dict):
                        item = find_key(v, key)
                        if item is not None: return item
                return None
            
            found_data = find_key(raw_json, 'adlisting')
            if found_data and 'data' in found_data and 'ads' in found_data['data']:
                ads_list = found_data['data']['ads']

        if not ads_list:
            print("❌ ERROR: Could not locate 'ads' list in the JSON.")
            # Debug: Print top-level keys to help user debug
            print(f"Available keys in props: {raw_json.get('props', {}).keys()}")
            return

        print(f"✅ Found {len(ads_list)} listings.")

        cleaned_data = []
        for ad in ads_list:
            item = {
                "listing_id": ad.get('ad_id'),
                "title": clean_text(ad.get('subject')),
                "description": clean_text(ad.get('body')),
                "price": ad.get('price'),
                "price_string": ad.get('price_string'),
                "area_m2": ad.get('size'),
                "address": clean_text(ad.get('street_name', 'Vinhomes Ocean Park')),
                "ward_name": ad.get('ward_name'),
                "district_name": ad.get('area_name'),
                "images": ad.get('images', []),
                "landlord": {
                    "account_id": ad.get('account_id'),
                    "name": clean_text(ad.get('account_name')),
                    "avatar": ad.get('avatar')
                },
                "features": {
                    "rooms": ad.get('rooms', 1),
                    "toilets": ad.get('toilets', 1),
                    "floors": ad.get('floors'),
                    "property_type": ad.get('category_name')
                },
                "timestamp": ad.get('date')
            }
            cleaned_data.append(item)

        with open(OUTPUT_JSON, 'w', encoding='utf-8') as outfile:
            json.dump(cleaned_data, outfile, ensure_ascii=False, indent=4)

        print(f"SUCCESS! Data saved to: {OUTPUT_JSON}")

    except Exception as e:
        print(f"Error parsing JSON: {e}")

if __name__ == "__main__":
    extract_html_to_json()
```

```python
import json
import os
from bs4 import BeautifulSoup

# ==============================================================================
# CONFIGURATION
# ==============================================================================
INPUT_FILE = 'nhatot_page.html'
OUTPUT_JSON = 'trial.json'
TARGET_DISTRICT = "Huyện Gia Lâm"  # <--- The specific filter

def clean_text(text):
    if not text: return ""
    return text.strip()

def extract_filtered_json():
    if not os.path.exists(INPUT_FILE):
        print(f"Error: {INPUT_FILE} not found.")
        return

    print(f"Reading {INPUT_FILE}...")
    with open(INPUT_FILE, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    script_tag = soup.find('script', id='__NEXT_DATA__')
    if not script_tag:
        print("Error: Could not find __NEXT_DATA__ tag.")
        return

    try:
        raw_json = json.loads(script_tag.string)
        
        # Attempt to find the ads list
        ads_list = []
        # Try Path 1
        try:
            ads_list = raw_json['props']['pageProps']['initialProps']['adlisting']['data']['ads']
        except KeyError:
            pass
        
        # Try Path 2
        if not ads_list:
            try:
                ads_list = raw_json['props']['pageProps']['adlisting']['data']['ads']
            except KeyError:
                pass

        if not ads_list:
            print("❌ ERROR: Could not find any ads in the file.")
            return

        print(f"Found {len(ads_list)} total ads (including unrelated 'Featured' ads).")
        print(f"Filtering for: {TARGET_DISTRICT}...")

        cleaned_data = []
        
        for ad in ads_list:
            # --- FILTERING STEP ---
            current_district = ad.get('area_name', '')
            
            if current_district != TARGET_DISTRICT:
                # Skip this ad if it's not in the target district
                continue

            item = {
                "listing_id": ad.get('ad_id'),
                "title": clean_text(ad.get('subject')),
                "description": clean_text(ad.get('body')),
                "price": ad.get('price'),
                "price_string": ad.get('price_string'),
                "area_m2": ad.get('size'),
                "address": clean_text(ad.get('street_name', 'Vinhomes Ocean Park')),
                "ward_name": ad.get('ward_name'),
                "district_name": current_district, 
                "images": ad.get('images', []),
                "landlord": {
                    "account_id": ad.get('account_id'),
                    "name": clean_text(ad.get('account_name')),
                    "avatar": ad.get('avatar')
                },
                "features": {
                    "rooms": ad.get('rooms', 1),
                    "toilets": ad.get('toilets', 1),
                    "floors": ad.get('floors'),
                    "property_type": ad.get('category_name')
                },
                "timestamp": ad.get('date')
            }
            cleaned_data.append(item)

        # Save Result
        with open(OUTPUT_JSON, 'w', encoding='utf-8') as outfile:
            json.dump(cleaned_data, outfile, ensure_ascii=False, indent=4)

        print(f"✅ SUCCESS! Extracted {len(cleaned_data)} valid listings for '{TARGET_DISTRICT}'.")
        print(f"Data saved to: {OUTPUT_JSON}")

    except Exception as e:
        print(f"Error parsing JSON: {e}")

if __name__ == "__main__":
    extract_filtered_json()
```
